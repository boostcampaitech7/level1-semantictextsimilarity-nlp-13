```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# CSV 파일 읽기
df = pd.read_csv('aug_train.csv')

# 분포 시각화
plt.figure(figsize=(10, 6))
sns.histplot(df['label'], bins=20, kde=True)
plt.title('Label Distribution')
plt.xlabel('Label')
plt.ylabel('Frequency')
plt.show()
```

```python

# 0.5 간격으로 bin 설정
bins = [i * 0.5 for i in range(11)]  # 0.0부터 5.0까지 0.5 간격

# 분포 시각화
plt.figure(figsize=(10, 6))
sns.histplot(df['label'], bins=bins, kde=False)
plt.title('Label Distribution (0.5 Intervals)')
plt.xlabel('Label')
plt.ylabel('Frequency')
plt.xticks(bins)  # x축 간격 설정
plt.show()
```

```python
import pandas as pd

# CSV 파일 읽기
df = pd.read_csv('dev_output.csv')

# MSE 계산
df['mse'] = abs(df['label'] - df['target'])

# MSE 기준으로 정렬
sorted_df = df.sort_values(by='mse', ascending=False)

# 정렬된 데이터 저장
sorted_df.to_csv('sorted_dev_output_kykim.csv', index=False)
```
